```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

import os
from datetime import datetime

pd.options.display.max_columns = None
```

```python
path = './project_input/activities_with_zones.csv'
df = pd.read_csv(path)
df.head()
```

```python
dist_df = df[['start_date_local', 'distance']]
dist_df.head()
```

```python
dist_df.columns = ['date', 'distance']
dist_df.head()
form_dist_df = pd.DataFrame({'date': [], 'distance': []})
```

```python
form_dist_df['date'] = pd.to_datetime(dist_df['date'])
form_dist_df['date'].iloc[0]
```

```python
form_dist_df['distance'] = dist_df['distance'].apply(lambda x : x / 1000)
form_dist_df['distance'].iloc[0]
```

```python
form_dist_df.head()
```

```python
form_dist_df.tail()
```

```python
plt.figure(figsize=(15,4), dpi=100)
plt.plot(form_dist_df['date'], form_dist_df['distance'], color='tab:red')
plt.gca().set(title='Distance per ride from 2013 to 2023', xlabel='Date', ylabel='Distance')
plt.show()
```

```python
form_dist_df.describe()
```

```python
df_2023 = form_dist_df[form_dist_df['date'].dt.year == 2023]
df_2023.describe()
```

```python
plt.figure(figsize=(15,4), dpi=100)
plt.plot(df_2023['date'], df_2023['distance'], color='tab:orange')
plt.gca().set(title='Distance per ride from 2023', xlabel='Date', ylabel='Distance')
plt.show()
```

```python
df_weekly_23 = df_2023.copy()
df_weekly_23['week'] = df_2023['date'].dt.isocalendar().week
df_grouped_weekly = df_weekly_23.groupby('week')
df_aggregated = df_grouped_weekly['distance'].sum().to_frame()
df_aggregated['week'] = df_aggregated.index
df_aggregated.head()
```

```python
plt.figure(figsize=(15,4), dpi=100)
plt.plot(df_aggregated['week'], df_aggregated['distance'])
plt.plot(df_aggregated['week'], df_aggregated['distance'], marker='o', linestyle=None)
plt.gca().set(title='Distance per week from 2023', xlabel='Date', ylabel='Distance')
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from dateutil.parser import parse

multiplicative_decomposition = seasonal_decompose(df_aggregated['distance'], model='multiplicative', period=5)

additive_decomposition = seasonal_decompose(df_aggregated['distance'], model='additive', period=5)

plt.rcParams.update({'figure.figsize': (16, 7)})
multiplicative_decomposition.plot().suptitle('Multiplicative Decomposition', fontsize=16)
plt.tight_layout()

additive_decomposition.plot().suptitle('Additive Decomposition', fontsize=16)
plt.tight_layout()

plt.show()
```

```python
# Using statmodels: Subtracting the Trend Component
from statsmodels.tsa.seasonal import seasonal_decompose

result_mul = seasonal_decompose(df_aggregated['distance'], model='multiplicative', period=5)
detrended = df_aggregated['distance'].values - result_mul.trend
plt.rcParams.update({'figure.figsize': (16, 4)})
plt.plot(detrended)
plt.title('Distance detrended by subtracting the trend component', fontsize=16)
```

```python
# Subtracting the Trend Component

# Time Series Decomposition
result_mul_des = seasonal_decompose(df_aggregated['distance'], model='multiplicative', period=5)

# Deseasonalize
deseasonalized = df_aggregated['distance'].values / result_mul_des.seasonal

# Plot
plt.plot(deseasonalized)
plt.title('Distance Deseasonalized', fontsize=16)
plt.plot()
```

```python
from statsmodels.tsa.stattools import adfuller
def adf_test(dataset):
     dftest = adfuller(dataset, autolag = 'AIC')
     print("1. ADF : ",dftest[0])
     print("2. P-Value : ", dftest[1])
     print("3. Num Of Lags : ", dftest[2])
     print("4. Num Of Observations Used For ADF Regression:",      dftest[3])
     print("5. Critical Values :")
     for key, val in dftest[4].items():
         print("\t",key, ": ", val)
adf_test(df_aggregated['distance'])
```

```python
df_differenced = df_aggregated.copy()
df_differenced['distance_diff'] = df_aggregated['distance'] - df_aggregated['distance'].shift(1)
df_differenced = df_differenced.dropna()
df_differenced.plot()
```

```python
adf_test(df_differenced['distance_diff'])
```

```python
from pmdarima import auto_arima
stepwise_fit = auto_arima(df_differenced['distance'], trace=True,
suppress_warnings=True)
```

```python
from statsmodels.tsa.arima.model import ARIMA
model=ARIMA(df_differenced['distance'],order=(5, 0, 4))
model=model.fit()
model.summary()
```

```python
start=len(df_differenced)
end=len(df_differenced)+12
pred=model.predict(start=start,end=end,typ='levels').rename('ARIMA Predictions')
pred.plot()
```

```python
df_differenced.tail()
```

```python
distances = df_differenced['distance']
distances.head()
```

```python
ser_pred = pred.copy()
ser_pred.index = range(43, 43 + 13)
ser_pred.tail()
```

```python
df_ser_pred = ser_pred.to_frame()
df_ser_pred['week'] = ser_pred.index
df_ser_pred.columns = ['distance', 'week']
df_ser_pred.head()
```

```python
print(df_ser_pred.shape)
last_row = df_differenced.iloc[-1][['distance', 'week']]
df_ser_pred.loc[last_row['week']] = last_row 
df_ser_pred = df_ser_pred.sort_index().reset_index(drop=True)
print(df_ser_pred.shape)
df_ser_pred.tail()
```

```python
# Plot the line graph
plt.figure(figsize=(10, 6))
plt.plot(df_differenced['week'], df_differenced['distance'], marker='o', linestyle='-', color='red')
plt.plot(df_ser_pred['week'], df_ser_pred['distance'], marker='o', linestyle='--', color='blue')

# Customize the plot
plt.title('Weekly Distance')
plt.xlabel('Week Number')
plt.ylabel('Distance')
plt.grid(True)

# Show a legend for the different colors
plt.legend(['< Week 43', 'Prediction: Week 43 and above'])

# Display the plot
plt.show()
```
